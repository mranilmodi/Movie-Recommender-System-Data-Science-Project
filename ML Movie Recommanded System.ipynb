### Data Preprocessing / Feature Engeenring

```python
# let's select the important feature to make recomandaded syatem
 
  # genres 
  # overview
  # popularity
  # title
  # movie_id
  # cast
  # crew

# these are important columns we should use in make system

movies = movies[['movie_id', 'title', 'overview','keywords', 'genres', 'cast', 'crew',  ]]
movies.head(2)
```

```python
# Let's create a link in cast columns /modify this columns
```

```python
# First problems is - values string[list[dict]]
# so we first of all string list convert into a list

import ast

def convert(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
    
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head(2)
```

```python
movies['keywords'] =movies['keywords'].apply(convert)
movies.head(2)
```

```python
def fetch_name(obj):
    l = []
    c = 0
    for i in ast.literal_eval(obj):
        if c <=3:
            l.append(i['name'])
        c = c +1
    return l
```

```python
# take 4 actors
movies['cast'] = movies['cast'].apply(fetch_name)
movies.head(2)
```

```python
def frtch_director(obj):
    l = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            l.append(i['name'])
    return l
```

```python
movies['crew'] = movies['crew'].apply(frtch_director)
movies.head(2)
```

```python
l = [] 
for i in range(movies['overview'].shape[0]):
    l.append(str(movies['overview'][i]).split())
```

```python
movies['overview'] = l
```

```python
movies.head()
```

```python
# Remove spce between tha word

def space(x):
    l = []
    for i in x:
        l.append(i.replace(" ",""))
    return l
```

```python
movies['keywords'] = movies['keywords'].apply(space)
movies['genres'] = movies['genres'].apply(space)
movies['cast'] = movies['cast'].apply(space)
movies['crew'] = movies['crew'].apply(space)
```

```python
movies.head(3)
```

```python
movies['tags'] = movies['overview'] + movies['keywords'] + movies['genres']  + movies['cast']  + movies['crew'] 
```

```python
new_df = movies[['movie_id', 'title', 'tags']]
new_df.head(10)
```

```python
# now convert list to string of tags
new_df['tags'] = new_df['tags'].apply(lambda x :" ".join(x))
new_df.head()
```

```python
new_df['tags'] =new_df['tags'].apply(lambda x :x.lower())
```

```python
new_df.head()
```

```python
import nltk

from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)  # ise list string me chnage ho jati he 
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

# Let's create model

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english') # isme sbse maximum 5000 work lega or unko column bnayeva
                                                              # jisme aaya use 1 or nhi 0
```

```python
vactor = cv.fit_transform(new_df['tags']).toarray()
```

```python
## Now find distance one movie to another movie

## jitna similirity jyada hogi utni distance km hogi
## or jitni distance jyada he utni similirity kmhogi
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vactor)
```

#### let's create function recomanded

```python
def recommanded(movie):
    index = new_df[new_df['title'] == movie].index[0]
    movie_list = sorted(list(enumerate(similarity[index])), reverse = True, key = lambda x: x[1])[0:6]
    
    for i in range(0,6):
        ind = movie_list[i][0]
        print(new_df['title'][ind])
        



```
